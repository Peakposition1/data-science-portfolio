# Superstore Sales Analysis

## Project Overview - This project analyzes the Superstore sales dataset to understand sales performance, profitability and customer trends. The goal is to identify insights that can help improve business decision-making.

## Dataset Description - The Superstore dataset contains sales records including order details, customer information, product categories, sales, profit and geographic data.

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt

# Loading data
df = pd.read_csv(r"C:\Users\Txkxllx\Desktop\Sample - Superstore.csv")

# Previewing data
df.head()
```

## Data Preparation - This dataset was already clean and required minimal preprocessing. Basic checks were performed to confirm data types and ensure there were no missing or invalid values affecting the analysis. Additional features, such as revenue calculations and date formatting, were created to support exploratory and visual analysis.

```python
# Info check

df.info()
```

## Sales Performance Analysis

```python
# Total Sales by Region
## Group the data by region and calculate the total Sales.

region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

# Visualization

```python
region_sales.plot(kind='bar',figsize=(8,5), title='Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

# Insight:
The West region has the highest total sales, while the South has the lowest. This shows that sales performance varies by region, which may require targeted marketing or regional sales strategies.

## Profitability Analysis

```python
# Total Profit by Region
## Group the data by region and calculate the total Profit.

region_profit = df.groupby('Region')['Profit'].sum()
region_profit
```

```python
# Visualize total profit by region

region_profit.plot(kind='bar',figsize=(8,5), color='green', title='Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.show()
```

# Insight:
The West region also has the highest total profit, similar to its sales performance, while the Central region has the lowest profit. This highlights that the regions with higher sales are contributing more to overall profit, which could inform regional strategies.

```python
# Total Sales by Category
## Group the data by category and calculate total sales.

category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
# Visualize total sales by category

category_sales.plot(kind='bar', figsize=(8, 5), color='orange', title='Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

# Insight: 
Technology has the highest total sales among categories, followed by Furniture and Office Supplies. This indicates which categories contribute most to revenue and can guide inventory and maketing priorities.

# Recommendation
-  Prioritizing inventory, marketing and sales efforts in the West and East regions, as they generate the highest total sales and offer the strongest revenue opportunities.
- Investigate the South region's lower performance to identify potential issues such as limited product availabity, weaker demand or pricing challenges and introduce targeted promotions to stimulate sales.
- Address low-profit or loss-making segments through pricing or cost optimization for the South and Central region.

The Superstar dataset analysis revealed key trends in sales and profitability across products, customers and regions. These insights can support data-driven decisions to improve overall business performance.
